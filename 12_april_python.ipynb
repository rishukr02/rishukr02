```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "#6EFF33", linestyle = "dashdot")
plt.plot(batsman["index"],batsman["RG Sharma"],color = "orange",linestyle = "dashdot")
plt.title("Rohit sharma vs virat kohli comparision  in ipl runs")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "#6EFF33",linestyle = "dashdot",linewidth = 2,marker ="D")
plt.plot(batsman["index"],batsman["RG Sharma"],color = "orange",linewidth =2,marker = "D")
plt.title("Rohit sharma vs virat kohli comparision  in ipl runs")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "orange",label ="virat")
plt.plot(batsman["index"],batsman["RG Sharma"],color = "red",label = "rohit")
plt.title("rohit sharma vs virat kohli comparison in ipl runs")
plt.xlabel("ipl runs")
plt.ylabel("runs scored")
plt.legend()
```
